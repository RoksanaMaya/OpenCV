# Import Libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

# Image read

```python
path='phoenix.jpg'
image_cv2 = cv2.imread(path)
image_plt=plt.imread(path)
```

# Displaying Pixels

```python
image_pixels=np.array(image_plt)
image_pixels
```

# Display

```python


# Display the image using Matplotlib
# cv2.imread has the colour map as BGR so it represents color differently
plt.subplot(1, 2, 1)
plt.imshow(image_cv2)
plt.title("BGR: CV2")
plt.axis('off')
# plt.imread represents color map as RGB
plt.subplot(1, 2, 2)
plt.imshow(image_plt)
plt.title("RGB: PLT")
plt.axis('off')  # Turn off axis labels

image_cv2.shape
```

# Color Conversion

```python
# Convert BGR to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off') 
```

```python
# Convert BGR to Grayscale
gray_image = cv2.cvtColor(image_cv2, cv2.COLOR_BGR2GRAY)

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Grayscale Image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

# Pixels of original and grayscale image

```python
image_pixels=np.array(image_plt)
print('Original image pixel Matrix', image_pixels.shape,image_pixels)
```

```python
gray_pixels=np.array(gray_image)
print('Grayscale image pixel Matrix', gray_pixels.shape,gray_pixels)
```

```python
# Split the image into color channels
blue, green, red = cv2.split(image_cv2)

# Increase intensity of the red channel
red = np.clip(red + 50, 50, 175)  # Increase intensity by 50 (clipped to stay within [0, 255])

# # Increase intensity of the red channel
#blue = np.clip(blue + 209, 157,  100)

# # Increase intensity of the red channel
# green = np.clip(green + 209, 100, 200)
# # Merge the modified channels back into a colored image
colored_image = cv2.merge([blue, green, red])

# # Convert BGR to RGB (OpenCV uses BGR by default)
colored_image_rgb = cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB)

# Display the original and colored images side by side

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Colored Image
plt.subplot(1, 2, 2)
plt.imshow(colored_image_rgb)
plt.title('Colored Image')
plt.axis('off')
# show
plt.show()
```

```python
# Convert BGR to RGB
pixel_value_rgb = cv2.cvtColor(image_cv2, cv2.COLOR_BGR2RGB)
print(f'RGB values at position (100, 100): {pixel_value_rgb[100, 100]}')
plt.imshow(pixel_value_rgb )
plt.axis('off')
```

# Scaling and Resize

```python
# Specify the target dimensions in pixels
target_width = 500
target_height = 800
resize_image=cv2.resize(image_rgb,(target_width,target_height))
plt.imshow(resize_image)
plt.axis('off')
resize_image.shape
```

```python
# reduce image by percentage
resize_image=cv2.resize(image_rgb, None, fx=.25, fy=.25)
plt.imshow(resize_image)
plt.axis('off')
image_rgb.shape, resize_image.shape
```

```python
# increase image by percentage
resize_image=cv2.resize(image_rgb, None, fx=2, fy=2)
plt.imshow(resize_image)
plt.axis('off')
image_rgb.shape, resize_image.shape
```

```python
# increase image by percentage another methos
percentage=2
target_width = int(image_rgb.shape[1]*percentage)
target_height = int(image_rgb.shape[0]*percentage)
resize_image=cv2.resize(image_rgb,(target_width,target_height))
plt.imshow(resize_image)
plt.axis('off')
image_rgb.shape, resize_image.shape
```
