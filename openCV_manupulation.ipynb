```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
path='phoenix.jpg'
image = cv2.imread(path)
```

```python
# Convert BGR to RGB (OpenCV uses BGR by default)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image using Matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Turn off axis labels

image_rgb.shape
```

```python
# Specify the target dimensions in pixels
target_width = 500
target_height = 800
resize_image=cv2.resize(image_rgb,(target_width,target_height))
```

```python
plt.imshow(resize_image)
plt.axis('off')
resize_image.shape
```

```python
# Specify the target dimensions in percentage
percentage=3
target_width = int(image_rgb.shape[1]*percentage)
target_height = int(image_rgb.shape[0]*percentage)
resize_image=cv2.resize(image_rgb,(target_width,target_height))
plt.imshow(resize_image)
plt.axis('off')
resize_image.shape
```

```python
# Convert BGR to Grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray' )
plt.axis('off')
gray_image.shape
```

```python



# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Grayscale Image
plt.subplot(1, 2, 2)
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
# Split the image into color channels
blue, green, red = cv2.split(image)

# Example: Increase intensity of the red channel
red = np.clip(red + 50, 0, 255)  # Increase intensity by 50 (clipped to stay within [0, 255])

# Example: Increase intensity of the red channel
blue = np.clip(blue + 150, 0, 255)

# Example: Increase intensity of the red channel
green = np.clip(green + 50, 0, 255)
# Merge the modified channels back into a colored image
colored_image = cv2.merge([blue, green, red])

# Convert BGR to RGB (OpenCV uses BGR by default)
colored_image_rgb = cv2.cvtColor(colored_image, cv2.COLOR_BGR2RGB)

# Display the original and colored images side by side
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Colored Image
plt.subplot(1, 2, 2)
plt.imshow(colored_image_rgb)
plt.title('Colored Image')
plt.axis('off')
# show
plt.show()
```
